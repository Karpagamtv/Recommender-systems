```python
import numpy as np
import pandas as pd
```

```python
# Load user ratings
df = pd.read_csv('G:/material/Python/Project/ml-1m/ml-1m/ratings.dat',sep='::', encoding='latin-1',
                 names=['UserID','MovieID','Rating','Timestamp'])
```

```python
# Load movie titles
movies_df = pd.read_csv("G:/material/Python/Project/Final_movie_after genres_0.2.csv",encoding='latin-1')
```

```python
df.head()
```

```python
movies_df.head()
```

```python
# Grouping and Ranking Data
rating = pd.DataFrame(df.groupby('MovieID')['Rating'].mean())
rating.head()
```

```python
rating['rating_count'] = pd.DataFrame(df.groupby('MovieID')['Rating'].count())
rating.head()
```

```python
rating.sort_values('rating_count', ascending=False).head()
```

```python
movies_df[movies_df['movieID']==50]
```

```python
# Preparing for Data Analysis
#user by Item utility matrix
movies_crosstab = pd.pivot_table(data=df, values='Rating', index='UserID', columns='MovieID')
movies_crosstab.head()
```

```python
AmericanBeauty_ratings = movies_crosstab[50]
AmericanBeauty_ratings[AmericanBeauty_ratings>=0]
```

```python
#Evaluating Similarity Based on Correlation
# corrwith method is used to find the correlation
similar_to_AmericanBeauty = movies_crosstab.corrwith(AmericanBeauty_ratings)

corr_AB = pd.DataFrame(similar_to_AmericanBeauty, columns=['PearsonR'])
# dropping null values
corr_AB.dropna(inplace=True)
corr_AB.head()
```

```python
AB_corr_summary = corr_AB.join(rating['rating_count'])
```

```python
# well correlated but less rating count
recc_movies = AB_corr_summary[AB_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False).head(10)
recc_movies['MovieID']= recc_movies.index
recc_movies
```

```python
type(recc_movies)
```

```python
summary = pd.merge(recc_movies, movies_df,on='MovieID')
summary[['MovieID','Title','Genres','directorName']]
```
